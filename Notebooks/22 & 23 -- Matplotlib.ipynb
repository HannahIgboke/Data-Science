# Matplotlib

## Anatomy of a figure
Matplotlib graphs data on figures - windows, jupyter widgets - each of which contain one or more axes. The following are parts of a figure:

- Figure: the whole figure
- Axes: this is an artist attached to a figure that contains a region for plotting data, and usually two/three axis objects that provide ticks and tick labels to provide scales for the data in the axes. Each axes also has a title, xlabel, and a ylabel.

P.S. Note the difference between Axes and Axis.

- Artist: basically, everything visible on the figure is an artist (including figure, axes and axis objects).  This includes Text objects, Line2D objects, collection objects, patch objects, etc. When the figure is rendered, all of the artists are drawn to the canvas. Most artists are tied to an axes and such an artist cannot be shared by multiple axes or moved from one to another.

### Annotating the plots(texts and arrows)

This can be done in two ways: plt.text() and plt.annotate(). The latter offers greater flexibility than the former.

```python
plt.figure(figsize=(4,3))

plt.plot(data1, data2, 'bo', label='scatter')
plt.plot(data1, data2, 'k-', label = 'line')

#Labeling the plot
plt.title('Line plot')
plt.xlabel('X_label')
plt.ylabel('Y_label')

#annotating the plot
plt.text(s='This is a text', x=0.5, y=0.7)

#Putting limits to x and y axis
plt.xlim([0,1])
plt.ylim([0,1])

#display the legend
plt.legend()

plt.show()
```

### Legend location

```python
plt.figure(figsize =(5, 3), layout = 'constrained')

plt.plot(data1, data1, label='linear', color ='blue')
plt.plot(data1, data1**2, label = 'quadratic', color = 'red')
plt.plot(data1, data1**3, label = 'cubic', color ='orange')

plt.title('Line/Quad/Cubic plots')
plt.xlabel('x_label')
plt.ylabel('y_label')

plt.text(s = 'Cubic', x=1.75, y=5.2)
plt.text(s='Quadratic', x=1.75, y=3.5)
plt.text(s='Linear', x =1.75, y=2)

plt.legend(bbox_to_anchor=(1,1)) #outside the plot box/area

plt.show()
```

### Labeling and annotating subplots

```python
fig, ax = plt.subplots(figsize=(3,3), facecolor='green')
fig.suptitle('Figure title')

#adding Axes title
ax.set_title('Axes 1 title', loc = 'left', fontsize='small')

#add axes grid
ax.grid(True)

#add axes face color
ax.set_facecolor('yellow')

#adding annotations
ax.annotate('Text', xytext=(0.2, 0.2), ha ='center',xy=(0.5, 0.5), arrowprops={})

plt.show()
```

```python
#lets repeat this but for a figure with two subplots

fig, ax = plt.subplots(2,2, figsize=(3,3), facecolor='green', layout ='constrained')
fig.suptitle('Figure title')

#add axes face color
ax.set_facecolor('yellow')

#adding annotations
ax.annotate('Text', xytext=(0.2, 0.2), ha ='center',xy=(0.5, 0.5), arrowprops={})

plt.show()
```

```python
#the face color was not set across all the subplots, we have to use a for loop to iterate and apply the color
fig, axs = plt.subplots(2,2, figsize=(3,3), facecolor='green', layout ='constrained')
fig.suptitle('Figure title')

#since axs is a numpy.ndarray, we have to iterate and make the changes to each plot
#remember ravel() fuunction in numpy?
#we use it to flatten the array so we can iterate through it

for ax in axs.ravel():
    ax.set_facecolor('yellow')
    
plt.show()
```

```python
#taking the above a step further

fig, axs = plt.subplots(2,2, figsize=(5,5), facecolor='green', layout='constrained')
fig.suptitle('Figure Title')

axes_title = ['Axes_1', 'Axes_2', 'Axes_3', 'Axes_4']
axes_color = ['yellow', 'purple', 'brown', 'orange']

#applying for each subplot

for idx, ax in enumerate(axs.ravel()):
    ax.set_facecolor(axes_color[idx])
    ax.set_title(axes_title[idx], loc ='center', fontsize='medium')
    ax.set_xlabel('x_label')
    ax.set_ylabel('y_label')
    
plt.show()
```

```python
#using more subplots

fig, axs = plt.subplots(1,3, figsize=(10,3), layout='constrained')
fig.suptitle('Comparing plots')

data = [data1, data1**2, data1**3]
labels = ['linear', 'quadratic', 'cubic']
colors = ['blue', 'brown', 'orange']
axes_title = ['Linear', 'Quadratic', 'Cubic']

for idx, ax in enumerate(axs.ravel()):
    ax.plot(data[0], data[idx], 
            label=labels[idx], 
            color=colors[idx])
    ax.axhline(1, ls='--', color='red')
    ax.axvline(1, ls='--', color = 'black')
               
    ax.set_title(axes_title[idx], loc = 'center', fontsize=10)
    ax.set_xlabel('x_label')
    ax.set_ylabel('y_label')
    ax.legend()
               
plt.show()
```

## Using Helper methods

```python
def my_plotter(axes, data1, data2, kind):
    if kind == 'line':
        out = axes.plot(data1, data2)
    else:
        out = axes.scatter(data1, data2)
    return out
```

```python
#using the helper function

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(5,3))
my_plotter(ax1, data1, data2, 'line')
my_plotter(ax2, data3, data4, 'scatter')

plt.show()
```

## Creating basic matplotlib plots

```python
np.random.seed(30)
data1=np.linspace(0,100, 10000)
data2=np.random.randint(1,6, size = (10000,))
data3, data4 = np.random.randn(2,10000)

df = pd.DataFrame({
    'col1': data1,
    'col2': data2,
    'col3': data3,
    'col4': data4
})

df.head()
```

```python
### Univariate plotting for numerical columns

fig, axs = plt.subplots(1,2, figsize=(6,3), layout = 'constrained')
fig.suptitle('Univariate plotting')

axs[0].hist(df['col3'])
axs[0].set_title('Histogram plot', fontsize=10)

axs[1].boxplot(df['col3'])
axs[1].set_title('Box plot')

plt.show()
```

```python
### Univariate plotting for categorical columns

fig, axs = plt.subplots(1,2, figsize=(6,3), layout ='constrained')
fig.suptitle('Univariate plotting - Categorical features')

bar_data = df['col2'].value_counts()
axs[0].bar(bar_data.index, bar_data.values)
axs[0].set_title('Bar plot', fontsize=10)

pie_data = df['col2'].value_counts(normalize=True)
axs[1].pie(pie_data.values, labels = pie_data.index, autopct='%1.1f%%')
axs[1].set_title('Pie chart', fontsize='medium')

plt.show()
```

```python
### Bivariate plotting for categorical columns

fig, axs = plt.subplots(1,3, figsize=(10,3), layout ='constrained')
fig.suptitle('Bivariate plotting - Categorical features')

axs[0].scatter(df['col3'], df['col4'])
axs[0].set_title('Scatter plot', fontsize='medium')

hexbin1 = axs[1].hexbin(df['col3'], df['col4'])
axs[1].set_title('Hexbin plot', fontsize='medium')
fig.colorbar(hexbin1, ax=axs[1])

hexbin2 = axs[2].hexbin(df['col3'], df['col4'], bins = 'log')
axs[2].set_title('Hexbin plot with log color scale')
fig.colorbar(hexbin2, ax = axs[2])

plt.show()
```
